# Bloch sphere

Review the properties of the Bloch sphere: https://www.st-andrews.ac.uk/physics/quvis/simulations_html5/sims/blochsphere/blochsphere.html 

https://qiskit.org/textbook/ch-states/representing-qubit-states.html

Example: $\frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$

### Cartesian Coordinates

```python
coords_cartesian = [1, 0, 0] # [x, y, z]
plot_bloch_vector(coords_cartesian, coord_type='cartesian', title="Cartesian coordinates")
```

### Spherical Coordinates

```python
coords_spherical = [1, np.pi/2, 0] # [Radius, Theta, Phi]
plot_bloch_vector(coords_spherical, coord_type='spherical', title="Spherical coordinates")
```
