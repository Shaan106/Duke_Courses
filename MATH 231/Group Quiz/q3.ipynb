```python
import math 
from math import comb
import scipy.integrate as integrate
from scipy.stats import binom
from scipy.stats import norm
from IPython.display import Markdown # Needed for the Markdown
from tabulate import tabulate
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lmbda = 2
a = 3.8

for i in range(4):
    X = []
    rv = np.random.poisson(lmbda*a)
    for j in range(rv):
        X.append(a*rnd.random())
    plt.figure(figsize=(8, 0.5))
    plt.scatter(X,[0 for k in X])
    plt.xlim([0,a])
    plt.yticks([])
    plt.ylabel(None)
    plt.xlabel("$x$")
    plt.ylim([-.02, 0.1])
    plt.title(f"Realization {i+1}")
    plt.show()
        
plt.tight_layout()
```

```python
lmbda = 2
b = 1.95

for i in range(4):
    X = []
    Y=[]
    rv = np.random.poisson(lmbda*(b**2))
    for j in range(rv):
        X.append(b*rnd.random())
        Y.append(b*rnd.random())
    plt.figure(figsize=(4, 4))
    plt.scatter(X,Y)
    plt.xlim([0,b])
    plt.ylabel(None)
    plt.xlabel("$x$")
    plt.ylim([0, b])
    plt.title(f"Realization {i+1}")
    plt.show()
        
plt.tight_layout()
```

```python
lmbda = 2
a = 3.8
N=10000

def findVal(lst):
    if len(lst) < 2:
        return None
    sorted_lst = sorted(lst)
    return sorted_lst[1]
    

def f(x):
    return math.exp(-2*x)*4*x

secondClosest = []
for i in range(N):
    distance = []
    rv = np.random.poisson(lmbda*a)
    for j in range(rv):
        distance.append(a*rnd.random())
    val = findVal(distance)
    if (val is not None):
        secondClosest.append(val)

X = np.linspace(0, max(secondClosest), 1000)
Y =[f(x) for x in X]

plt.hist(secondClosest, bins=30, rwidth = 0.9, density=True)
plt.plot(X, Y)
plt.xlabel("Second Closest Distance to Origin")
plt.title("Histogram of Second Closest Distance in 1D Poisson Process")
plt.show()
```
