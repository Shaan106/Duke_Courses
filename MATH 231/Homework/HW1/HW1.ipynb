```python
import matplotlib.pyplot as plt
```

```python
#should be 0,7,14,21,28,35,42,49 = 8/50 = approx 0.16

N =5000
samples=[]
divisibleOverTime=[]
divisible=0
for i in range(N):
  sample=drawRandomly0To49()
  samples.append(sample)
  if ((sample % 7)==0):
    divisible+=1

  # Need to add 1 to i because it starts at 0
  divisibleOverTime.append(divisible/(i+1)) 
#print out frequency
print(f"{divisible} of sample divisible by 7 in {N} samples.")
print("The observed fraction divisible is {:.4g}".format(divisible/N))
#plots
plt.plot(range(1,N+1),divisibleOverTime)
plt.xlabel("Number of Samples")
plt.ylabel("Observe frequence of divisiblity")
plt.title("Probablity is divisible by 7")
plt.show()
```
